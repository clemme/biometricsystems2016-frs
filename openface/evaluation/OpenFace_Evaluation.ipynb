# Evaluate Probe Results

```python
# Load probe results
templates = pd.DataFrame.from_csv('../probe/lfw_probing_results.csv')
templates = templates[templates['Correct'] != pd.isnull]
# Compute score (distance 0 = 100%, distance 4 = 0%)
templates['Score'] = templates.apply(lambda x: (100 - (100/4) * x['Distance']) / 100,axis=1)
templates.head()
```

```python
# Raw data display
size = len(templates)
correct = templates['Correct'].value_counts()
true_positive = correct.iloc[0]
true_negative = correct.iloc[1]
print "TP: {}, TN: {} out of {}".format(true_positive, true_negative, size)
#correct
```

```python
# Extract actual best match (if best match was correct)
actual = templates['Correct'].values.tolist()
# Extract score for best match
predictions = templates['Score'].values.tolist()

# Compute ROC + False Positive Rate + True Positive Rate + AUC
false_positive_rate, true_positive_rate, thresholds = roc_curve(actual, predictions)
roc_auc = auc(false_positive_rate, true_positive_rate)

# Display ROC curve
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate, true_positive_rate, 'b', label='OpenFace LFW AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.1])
plt.ylim([-0.1,1.1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```

```python
#http://stackoverflow.com/questions/31324218/scikit-learn-how-to-obtain-true-positive-true-negative-false-positive-and-fal
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
    for i in range(len(y_hat)): 
        if y_actual[i]==1 and y_actual!=y_hat[i]:
           FP += 1
    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==0:
           TN += 1
    for i in range(len(y_hat)): 
        if y_actual[i]==0 and y_actual!=y_hat[i]:
           FN += 1

    return(TP, FP, TN, FN)

pnames = templates['Probe_Name'].values.tolist()
bmnames = templates['Best_Match'].values.tolist()

print pnames[:10]
print bmnames[:10]

tp, fp, tn, fn = perf_measure(pnames, bmnames)

print "tp {}, fn {}, tn {}, fn {}".format(tp, fp, tn, fn)
```

```python
# Map names to numbers
dic = {}
i = 0
for names in [pnames, bmnames]:
    for name in names:
        if not name in dic:
            dic[name] = i
            i = i + 1
            
pactual = [dic[p] for p in pnames]
print "Length {} : {}...".format(len(pactual), pactual[:10])
bmactual = [dic[p] for p in bmnames]
print "Length {} : {}...".format(len(bmactual), bmactual[:10])

tp, fp, tn, fn = perf_measure(pactual, bmactual)

print "tp {}, fn {}, tn {}, fn {}".format(tp, fp, tn, fn)
```
