# Evaluate Probe Results

```python
# Load probe results
templates = pd.DataFrame.from_csv('../probe/lfw_probing_results.csv')
templates = templates[templates['Correct'] != pd.isnull]
# Compute score (distance 0 = 100%, distance 4 = 0%)
templates['Score'] = templates.apply(lambda x: (100 - (100/4) * x['Distance']) / 100,axis=1)

templates_meds = pd.DataFrame.from_csv('../probe/meds_probing_results.csv')
templates_meds = templates_meds[templates_meds['Correct'] != pd.isnull]
templates_meds['Score'] = templates_meds.apply(lambda x: (100 - (100/4) * x['Distance']) / 100,axis=1)


templates.head()
```

```python
# Raw data display
size = len(templates)
correct = templates['Correct'].value_counts()
true_positive = correct.iloc[0]
true_negative = correct.iloc[1]
print "Population: {}".format(size)
print "TP: {}, TN: {}".format(true_positive, true_negative)
```

```python
# LFW
# Extract actual best match (if best match was correct)
actual = templates['Correct'].values.tolist()
# Extract score for best match
predictions = templates['Score'].values.tolist()
# Compute ROC + False Positive Rate + True Positive Rate + AUC
false_positive_rate, true_positive_rate, thresholds = roc_curve(actual, predictions)
roc_auc = auc(false_positive_rate, true_positive_rate)

# MEDS
actual_meds = templates_meds['Correct'].values.tolist()
predictions_meds = templates_meds['Score'].values.tolist()
false_positive_rate_meds, true_positive_rate_meds, thresholds_meds = roc_curve(actual_meds, predictions_meds)
roc_auc_meds = auc(false_positive_rate_meds, true_positive_rate_meds)

from pylab import rcParams
rcParams['figure.figsize'] = 10, 7

# Display ROC curve
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate, true_positive_rate, 'r', label='OpenFace (Similarity) LFW AUC = %0.2f'% roc_auc)
plt.plot(false_positive_rate_meds, true_positive_rate_meds, 'b', label='OpenFace (Similarity) MEDS AUC = %0.2f'% roc_auc_meds)
plt.legend(loc='lower right')
#plt.plot([0,1],[0,1],'r--')
plt.xscale('log')
plt.xlim([-0.0,1.0])
plt.ylim([-0.0,1.0])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```

```python
TPR_int = [0.4, 0.5, 0.65, 0.75, 0.85, 0.95]
FPR_int = [1e-06, 1e-05, 1e-04, 1e-03, 1e-02, 1e-01]

fpr_at_tpr = zip(TPR_int, [np.interp(tpr_at, true_positive_rate, false_positive_rate) for tpr_at in TPR_int])
tpr_at_fpr = zip(FPR_int, [np.interp(fpr_at, false_positive_rate, true_positive_rate) for fpr_at in FPR_int])

print "Table of True Accept Rates at various False Accept Rates"
for (far, tar) in tpr_at_fpr:
    print "FAR = {0}\t{1:0.3f}".format(far, tar)
    
print "\nTable of False Accept Rates at various True Accept Rates"
for (tar, far) in fpr_at_tpr:
    print "TAR = {0}\t{1:0.3f}".format(tar, far)
```

## Additional Analysis

```python
# Translate names to labels (for confusion matrix)
actuals = templates['Probe_Name'].values.tolist()
preds = templates['Best_Match'].values.tolist()

name_to_label_dic = {}
label_to_name_dic = {}
i = 0
for names in [actuals, preds]:
    for name in names:
        if not name in name_to_label_dic:
            name_to_label_dic[name] = i
            label_to_name_dic[i] = name
            i = i + 1
            
l_actuals = [name_to_label_dic[p] for p in actuals]
print "[Actuals (Probes)] Length {} : {}...".format(len(l_actuals), l_actuals[:10])
l_preds = [name_to_label_dic[p] for p in preds]
print "[Predicted]        Length {} : {}...".format(len(l_preds), l_preds[:10])
```

```python
# https://pypi.python.org/pypi/pandas_confusion/
from pandas_confusion import ConfusionMatrix
# Build confusion matrix for analysis
y_actu = l_actuals
y_pred = l_preds

cm = ConfusionMatrix(y_actu, y_pred)
cm_stats = cm.stats()
```

```python
# Switch index and column names (for easier manipulation)
cm_stats_class_t = cm_stats['class'].transpose()

# All data subjects (labelled)
subjects = cm_stats_class_t.index
# Individual analysis stats
subject_stats = cm_stats_class_t.columns

tp_all = cm_stats_class_t['TP: True Positive'].sum()
tn_all = cm_stats_class_t['TN: True Negative'].sum()
fp_all = cm_stats_class_t['FP: False Positive'].sum()
fn_all = cm_stats_class_t['FN: False Negative'].sum()

print "Overall:"
print "TP = {}".format(tp_all)
print "TN = {}".format(tn_all)
print "FP = {}".format(fp_all)
print "FN = {}".format(fn_all)
```
