# ROC - Receiver Operating Characteristics

```python
roc_openface_lfw  = pd.DataFrame.from_csv('roc-openface-lfw.csv')
roc_openface_meds = pd.DataFrame.from_csv('roc-openface-meds.csv')
```

```python
ax = roc_openface_lfw.plot(x='x', kind='line', figsize=(9,6), title="Receiver Operating Curve", xlim=(-0.01,1.01), ylim=(0.0,1.01), logx=True)
ax2 = roc_openface_meds.plot(x='x', ax=ax)
ax2.set_xlabel('False Accept Rate')
ax2.set_ylabel('True Accept Rate')
ax2.xaxis.set_major_formatter(FormatStrFormatter('%2.3f'))
```
