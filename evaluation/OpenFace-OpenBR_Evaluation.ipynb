```python
from __future__ import division
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import matplotlib
import pandas as pd
import scipy
import numpy as np
from IPython.display import Image
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
pd.set_option("display.max_columns",30)
pd.set_option('max_colwidth', 256)

matplotlib.style.use('ggplot')
%matplotlib inline
```

# ROC - Receiver Operating Characteristics

```python
roc_openface_lfw  = pd.DataFrame.from_csv('./roc-openface-lfw.csv')
roc_openface_meds = pd.DataFrame.from_csv('./roc-openface-meds.csv')
roc_openbr = pd.DataFrame.from_csv('./OpenBR_ROC.csv')[['OpenBR 4SF-LFW AUC = 0.84', 'OpenBR 4SF-MEDS AUC = 0.95', 'X']]
```

```python
ax = roc_openface_lfw.plot(x='x', kind='line', figsize=(9,6), linewidth=2, title="Receiver Operating Curve", xlim=(-0.01,1.01), ylim=(0.0,1.01), logx=True)
ax2 = roc_openface_meds.plot(x='x', ax=ax, linewidth=2)
ax3 = roc_openbr.plot(x='X', ax=ax2, linewidth=2)
ax3.set_xlabel('False Accept Rate')
ax3.set_ylabel('True Accept Rate')
#ax3.xaxis.set_major_formatter(FormatStrFormatter('%2.6f'))
```

# CMC - Cumulative Match Characteristics

```python
cmc_openface_lfw = pd.DataFrame.from_csv('./cmc-openface-lfw.csv')
cmc_openface_meds = pd.DataFrame.from_csv('./cmc-openface-meds.csv') 
cmc_openbr = pd.DataFrame.from_csv('./OpenBR_CMC.csv')[['OpenBR 4SF-LFW', 'OpenBR 4SF-MEDS', 'x']]
```

```python
# CMC
ax1 = cmc_openface_lfw.plot(x='x', title='Cumulative Match Characteristics', linewidth=2, figsize=(9,6), logx=True, ylim=(0,1), xticks=[1, 25, 50, 100, 200])
ax2 = cmc_openface_meds.plot(x='x', ax=ax1, linewidth=2)
ax3 = cmc_openbr.plot(x='x', ax=ax2, linewidth=2)
ax3.set_xlabel('Rank')
ax3.set_ylabel('Identification Rate')
ax3.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
```
